```python
# 원티드 Crawling

import requests
import datetime

parsed_data = []

base_url = "https://www.wanted.co.kr/api/v4/jobs/"
parse_company = requests.get("https://www.wanted.co.kr/api/v4/jobs?1700544430517&country=kr&tag_type_ids=518&job_sort=job.latest_order&locations=all&years=-1&limit=100&offset=0").json()
recur_ids =[temp['id'] for temp in parse_company['data']]  # 채용공고 아이디
timestamp = int(datetime.datetime.now().timestamp())

for recur_id in recur_ids:
    url =f"{base_url}{recur_id}?{timestamp}"
    parsed_data=requests.get(url).json()

    company = parsed_data['job']['company']['name'] #회사 이름
    intro = parsed_data['job']['detail']['intro']  # 회사 소개
    
    location = parsed_data['job']['address']['full_location']  # 근무지역
    title = parsed_data['job']['position'] # 제목
    main_job = parsed_data['job']['detail']['main_tasks'] # 주요업무
    require = parsed_data['job']['detail']['requirements']  # 자격요건
    prefer_point = parsed_data['job']['detail']['preferred_points'] # 우대사항
    benefits = parsed_data['job']['detail']['benefits'] # 혜택 및 복지
    deadline = parsed_data['job']['due_time'] # 마감날짜 None=상시
    
    geo_location = parsed_data['job']['address'].get('geo_location')
    if(geo_location):
        lat = geo_location['n_location']['lat']  # 위도  (근무지역)
        lng = geo_location['n_location']['lng']  # 경도  (근무지역)  -> geo_location 없는경우 있음 if문처리
    else:
        lat = None
        lng = None
        
    print(company)
   
    
```
